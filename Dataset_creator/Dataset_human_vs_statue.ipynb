```python
#importing the libraries
from PIL import Image
import requests
from io import BytesIO
from torchvision import models
import torch
from torchvision import transforms
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
import random
import os 
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/content/drive/MyDrive/Vision/Human_dataset/Human_test'
```

```python
from torchvision import transforms

transform = transforms.Compose([transforms.Resize(256), transforms.CenterCrop(256), transforms.ToTensor()])
 #transforms.CenterCrop(224),
 #transforms.Normalize( mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])])

change = transforms.ColorJitter(brightness=(0.5,1.5), contrast=(1), saturation=(0.5,1.5), hue=(-0.1,0.1))
```

```python
#appending the pics to the training data list
training_data = []
for img in os.listdir(path):
  img = Image.open(os.path.join(path,img))
  img_t = transform(img)
  batch_t = torch.unsqueeze(img_t, 0)

  training_data.append(batch_t)
```

```python
torch.save(training_data, '/content/drive/MyDrive/Vision/Human_dataset/human_test.pt')
```
